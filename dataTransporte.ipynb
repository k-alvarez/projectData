```python
# Eliminar filas donde 'Ciudad' es igual a Bogotá
dt_sinbogota = dt[dt['Ciudad'] != 'Bogotá']
```

```python
pd.unique(dt_sinbogota['Ciudad'])
```

```python
# Definir la fecha específica
fecha_especifica = pd.to_datetime('2020-05-01')
```

```python
# Filtrar las filas que tienen una fecha mayor a la fecha específica
dt_filtrado = dt_sinbogota[dt_sinbogota['Fecha'] > fecha_especifica]
```

```python
dt_filtrado.head()
```

```python
dt_filtrado.describe()
```

```python
# Agrupar y calcular la media
dia = dt_filtrado.groupby(['Sistema', 'Ciudad'])['Pasajeros/dia'].mean()
dia = dia.sort_values(ascending=False)
print(dia)
```

```python
bogotasuacha=dt_filtrado[dt_filtrado['Ciudad']=='Bogotá y Soacha']
bogotasuacha.head()
bogysua_media=bogotasuacha['Pasajeros/dia'].mean()
plt.figure(figsize=(14, 6))
plt.plot(bogotasuacha['Fecha'], bogotasuacha['Pasajeros/dia'], marker='d', linestyle='-', color='#468faf', label='Dia')
#plt.plot(bogotasuacha['Fecha'], bogotasuacha['Pasajeros día típico laboral'], marker='o', linestyle='', color='aqua', label='Dia Laboral')
#plt.plot(bogotasuacha['Fecha'], bogotasuacha['Pasajeros día sábado'], marker='o', linestyle='', color='cadetblue', label='Sabado')
#plt.plot(bogotasuacha['Fecha'], bogotasuacha['Pasajeros día festivo'], marker='o', linestyle='', color='lightblue', label='Festivo')
plt.axhline(bogysua_media, color='red', linestyle='--', label=f'Media: {bogysua_media}')
plt.xlabel('Fecha (Mayo-Septiembre)',fontsize=10)
plt.ylabel('Pasajeros/Dia',fontsize=10)
plt.title('Pasajeros/Dia en Bogotá y Soacha',fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend()
#plt.grid(True)
plt.show()
```
